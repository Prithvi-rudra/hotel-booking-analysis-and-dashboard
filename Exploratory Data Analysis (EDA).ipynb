# Importing libraries and dataset

```python
import pandas as pd
import numpy as np

hb_dt = pd.read_csv("cleaned_hotel_bookings.csv")
hb_dt
```

# Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the cancellation counts
drop_counts = hb_dt['is_canceled'].value_counts()

# Create a bar plot
plt.figure(figsize=(6, 4))
sns.barplot(x=drop_counts.index, y=drop_counts.values)
plt.title('Dropping Distribution (0 = Not Canceled, 1 = Canceled)')
plt.ylabel('List of Bookings')
plt.show()
```

```python
# Comparing cancellation by hotel type

# Create a count plot separated by hotel type
sns.countplot(x='hotel', hue='is_canceled', data=hb_dt)

# Add titles
plt.title('Cancellations by Hotel Type')
plt.show()
```

# Impact of Price (ADR) on Cancellations

```python
# Create a box plot to compare ADR for canceled vs not canceled
plt.figure(figsize=(8, 5))
sns.boxplot(x='is_canceled', y='adr', data=hb_dt)

# Add titles
plt.title('Price (ADR) vs. Cancellation Status')
plt.xlabel('Status (0 = Not Canceled, 1 = Canceled)')
plt.ylabel('Average Daily Rate (ADR)')
plt.show()
```
